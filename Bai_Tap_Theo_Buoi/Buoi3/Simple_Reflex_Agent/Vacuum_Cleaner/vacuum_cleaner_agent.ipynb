Thiết lập tập luật cho bài toán máy hút bụi có vật cản
1. If  phong[x][y] hợp lệ thì có thể di chuyển 
2. If  phong[x][y] bẩn thì sẽ hút bụi 
3. If  phong[x][y] nếu là vật cản thì không thể di chuyển đến 
4. If  phong[x][y] sạch thì đi tìm ô khác 


```python
import random
phong = [["ban", "sach", "vat can", "sach"], 
         ["ban", "sach", "sach", "vat can" ],
         ["sach", "vat can", "vat can", "sach"],
         ["ban", "sach", "vat can", "sach"]]
#Vị trí máy hút bụi hiện tại 
x, y = 2, 0

#Tập luật 
def luat(x, y):
    if x < 0 or x > 3 or y < 0  or y > 3:
        return False
    if phong[x][y] == "ban":
        return True
    if phong[x][y] == "vat can":
        return False
    if phong[x][y] == "sach":
        return True


def agent_di_chuyen(x, y):
#Kiểm tra xem phòng đã sạch hay chưa 
    def kiem_tra():
        for r in range(len(phong)):
            for c in range(len(phong[0])):
                if phong[r][c] == "ban":
                    return False
        return True

#Máy hút bụi có thể di chuyển một số vị trí left - right - up - down 
    def di_chuyen(x, y):
        tap_di_chuyen = []
        if (y - 1 >= 0):
            tap_di_chuyen.append(["trai", [x, y - 1]])
        if (y + 1 <= 3):
            tap_di_chuyen.append(["phai", [x, y + 1]])
        if (x - 1 >= 0 ):
            tap_di_chuyen.append(["len", [x - 1, y]])
        if (x + 1 <= 3 ):
            tap_di_chuyen.append(["xuong", [x + 1, y]])

        return tap_di_chuyen
#Máy hút bụi có thể hút bụi 
    def hut_bui(x, y):
        phong[x][y] = "sach"

    di = "Bat dau: " 
    so_buoc = 0
    if phong[x][y] == "ban":
        hut_bui(x, y)
    while(kiem_tra() == False and so_buoc <= 100):
        tap_di_chuyen = di_chuyen(x, y)
        buoc_di_tiep = random.choice(tap_di_chuyen)
        if luat(buoc_di_tiep[1][0], buoc_di_tiep[1][1]) == True:
            if phong[buoc_di_tiep[1][0]][buoc_di_tiep[1][1]] == "ban":
                hut_bui(buoc_di_tiep[1][0], buoc_di_tiep[1][1])
            x, y = buoc_di_tiep[1][0], buoc_di_tiep[1][1]
            di += " ---> " + buoc_di_tiep[0]
        so_buoc += 1
        #Nếu là vật cản thì không làm gì, không đi đâu hết 
    if kiem_tra() == False:
        print("Khong the don nha xong!")
    else:
        print("Da don nha xong rui!")
        print("Huong di chuyen: ", di)
        
def main():
    agent_di_chuyen(x, y)
    for r in range(len(phong)):
        for c in range(len(phong[0])):
            print(phong[r][c] + " " ,end = "")
        print()   
main()




```
