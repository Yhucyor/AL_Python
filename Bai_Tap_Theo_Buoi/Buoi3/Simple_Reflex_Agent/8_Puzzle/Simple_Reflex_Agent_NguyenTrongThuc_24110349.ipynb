```python
#Source Code
#Xác định tập luật
#1. Xác định vị trí ô trống
#2. Xác định vị trí liền kề[ trái, phải, lên, xuống] mà ô trống có thể duy chuyển
#3. Hoán đổi ô số với vị trí

import random
#Matrix:
maxtrixPuzzle = [[0, 1, 2], [6, 7, 8], [9, 5, 4]]
blankCell = (0, 0)

#Print Matrix
def printMatrix():
  for r in range(3):
    for c in range(3):
      print(maxtrixPuzzle[r][c], end = " ")
      if c == 2:
        print()

#State - Determine position of the blank
def determineBlank():
  for r in range(3):
    for c in range(3):
      if maxtrixPuzzle[r][c] == 0:
        return (r, c)

#Rules - Blank Move - Triển khai tập luật
def rule(blankCell):
  x = blankCell[0]
  y = blankCell[1]
  move = []
  #left
  if(y -1 >= 0):
    move.append((x, y - 1))
  #right
  if(y + 1 < 3):
    move.append((x, y + 1))
  #up
  if(x - 1 >= 0):
    move.append((x - 1, y))
  #downs
  if(x + 1 < 3):
    move.append((x + 1, y))
  return move

#Rules + State
def solveMatrix():
  blankCell = determineBlank()
  move = rule(blankCell)
  randomMove = random.choice(move)
  #Tiến hành Swap
  print("Nuoc di cua Agent la: " + str(randomMove))
  tmp = maxtrixPuzzle[blankCell[0]][blankCell[1]]
  maxtrixPuzzle[blankCell[0]][blankCell[1]] = maxtrixPuzzle[randomMove[0]][randomMove[1]]
  maxtrixPuzzle[randomMove[0]][randomMove[1]] = tmp

#Main
def Test():
  global maxtrixPuzzle
  print("Nhap bang tro choi 8 puzzle: ")
  tmpMatrix = []
  for _ in range(3):
    r = list(map(int,input().split()))
    tmpMatrix.append(r)
  maxtrixPuzzle = tmpMatrix
  solveMatrix()
  printMatrix()

Test()
```

```python
#Sourcode
import random
#Matrix:
maxtrixPuzzle = [[0, 1, 2], [6, 7, 8], [9, 5, 4]]
blankCell = (0, 0)
matrixSolve = [[1,2, 3], [4, 5, 6], [7, 8, 0]]


#Print Matrix
def printMatrix():
  for r in range(3):
    for c in range(3):
      print(maxtrixPuzzle[r][c], end = " ")
      if c == 2:
        print()


#State - Determine position of the blank
def determineBlank():
  for r in range(3):
    for c in range(3):
      if maxtrixPuzzle[r][c] == 0:
        return (r, c)

#Rules - Blank Move - Triển khai tập luật
def rule(blankCell):
  x = blankCell[0]
  y = blankCell[1]
  move = []
  #left
  if(y -1 >= 0):
    move.append((x, y - 1))
  #right
  if(y + 1 < 3):
    move.append((x, y + 1))
  #up
  if(x - 1 >= 0):
    move.append((x - 1, y))
  #downs
  if(x + 1 < 3):
    move.append((x + 1, y))
  return move


#Rules + State
def solveMatrix():
  blankCell = determineBlank()
  stepMove = []
  #Lưu trạng thái cũ
  statedMatrix = set()
  statedMatrix.add(str(maxtrixPuzzle))
  numofMove = 0

  while (maxtrixPuzzle != matrixSolve and numofMove < 1000):
    move = rule(blankCell)
    randomMove = random.choice(move)

    #Tiến hành Swap
    tmp = maxtrixPuzzle[blankCell[0]][blankCell[1]]
    maxtrixPuzzle[blankCell[0]][blankCell[1]] = maxtrixPuzzle[randomMove[0]][randomMove[1]]
    maxtrixPuzzle[randomMove[0]][randomMove[1]] = tmp


    if str(maxtrixPuzzle) not in statedMatrix:
      statedMatrix.add(str(maxtrixPuzzle))
      stepMove.append(randomMove)
      blankCell = randomMove
    else:
          tmp = maxtrixPuzzle[blankCell[0]][blankCell[1]]
          maxtrixPuzzle[blankCell[0]][blankCell[1]] =  maxtrixPuzzle[randomMove[0]][randomMove[1]]
          maxtrixPuzzle[randomMove[0]][randomMove[1]] = tmp
    numofMove += 1
  #print("Nuoc di cua Agent de giai 8 -Puzzle: ", str(stepMove))
  if maxtrixPuzzle != matrixSolve :
    print("Khong the gia duoc!")
  else:
    print("Nuoc di cua Agent de giai 8 -Puzzle: ", str(stepMove))


#Main
def Test():
  global maxtrixPuzzle
  print("Nhap bang tro choi 8 puzzle: ")
  tmpMatrix = []
  for _ in range(3):
    r = list(map(int,input().split()))
    tmpMatrix.append(r)
  maxtrixPuzzle = tmpMatrix
  solveMatrix()
  printMatrix()

Test()
```
