Bài 2: Xây dựng AI Agent huấn luyện máy hút bụi có thể hút bụi (trường hợp không có vật cản)

Xây dựng tập luật máy hụt bụi
R1: Nếu ô hiện tại có bụi thì tiến hành Hút Bụi
R2: Di chuyển lên trên (Nếu không ở hàng trên cùng)
R3: Di chuyển xuống dưới (Nếu không ở hàng dưới cùng)
R4: Di chuyển sang trái (Nếu không ở cột cuối bên trái)
R5: Di chuyển sang phải (Nếu không ở cột cuối bên phải) 

```python
import random 
#Dirty = 1, CLean = 0 
phong = []
cx = 0
cy = 0
so_buoc = 0
#Lưu trạng các ô đã đi qua 
agent_model = set() 

def rule(x, y, actions):
    if phong[x][y] == 1:
        phong[x][y] = 0
    moves = []
    for i in range(len(actions)):
        if actions[i] == "Up": moves.append([x - 1, y])
        if actions[i] == "Down": moves.append([x + 1, y])
        if actions[i] == "Left": moves.append([x, y - 1])
        if actions[i] == "Right": moves.append([x, y + 1])
    return moves

#Model Based Agent 
def Agent_Based_AI(percept):
    global cx, cy
    #1. Update - State: Cap nhat trang thai 
    cx, cy = percept
    agent_model.add((cx, cy))

    moves = [] 
    def get_action(i, j):
        actions = []
        if i > 0: actions.append("Up")
        if i < 2: actions.append("Down")
        if j > 0: actions.append("Left")
        if j < 2: actions.append("Right")
        return actions
    moves = get_action(cx, cy)
    #Rule - Match (state, rule)
    actions = rule(cx, cy, moves)

    #2. Đưa ra quyêt định - Nếu đã đi qua ô đó thì không đi nữa 
    unvisited_actions = []
    for di_chuyen in actions:
        if tuple(di_chuyen) not in agent_model:
            unvisited_actions.append(di_chuyen)
    if len(unvisited_actions):
        return random.choice(unvisited_actions)
        
    return random.choice(actions)

def kiem_tra():
    for r in range(len(phong)):
        for c in range(len(phong[r])):
            if phong[r][c] == 1:
                return False
    return True

def main():
    print("Nhap ma tran 3x3 moi: ")
    tmpPhong = []
    for _ in range(3):
        r = list(map(int, input().split()))
        tmpPhong.append(r)
    global phong, cx, cy, so_buoc, agent_model
    phong = tmpPhong
    cx = random.randint(0, 2)
    cy = random.randint(0, 2)
    so_buoc = 0
    agent_model = set()
    #AI Agent - Enviromment 
    duong_di =  "(" + str(cx) + "," + str(cy) + ")"
    while(kiem_tra() == False and so_buoc <= 100):
        #Cung cấp từ môi trường
        percept = (cx, cy)
        #Agent AI đưa ra hành động
        di_chuyen = Agent_Based_AI(percept)
        cx, cy = di_chuyen 
        duong_di += "-->" + str(di_chuyen)
        so_buoc += 1

    if (kiem_tra() == True):
        print("Da lau nha xong rui!" + "So buoc di: " + str(so_buoc))
        print(duong_di)
    else:
        print("Lau nha chua xong!")

    for i in range(len(phong)):
        for j in range(len(phong[i])):
            print(phong[i][j], end = " ")
        print()

main() 
```
